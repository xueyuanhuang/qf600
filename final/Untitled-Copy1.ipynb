## minimum-variance frontier
Briefly explain the economic significance and relevance of the minimum-variance frontier to an investor.

**Minimum-variance frontier represents outermost envelope of attainable portfolios ⇒ no attainable portfolio exists on the left of the minimum-variance frontier.**
>**Risk-Return Tradeoff**: The minimum-variance frontier helps investors understand the tradeoff between risk and return. It illustrates that, for a given level of expected return, there is an optimal portfolio with the lowest possible risk. Investors can use this information to make informed decisions about their risk tolerance and desired returns.

>**Portfolio Diversification**: The concept of the minimum-variance frontier highlights the benefits of diversification. By investing in a mix of assets with low correlations, investors can achieve a lower overall portfolio risk than by holding individual assets. This diversification can improve risk-adjusted returns.

>**Efficient Frontier**: The minimum-variance frontier is a key component of the efficient frontier, which shows all portfolios that provide the highest expected return for a given level of risk. Investors aim to construct portfolios that lie on or above the efficient frontier, as these portfolios offer the best risk-return tradeoff.

>**Portfolio Optimization**: Investors can use the minimum-variance frontier to optimize their portfolios by selecting a mix of assets that aligns with their risk preferences and return expectations. This optimization helps investors achieve the best possible risk-return balance based on their specific goals and constraints.

# 2.CAPM

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore")
```

```python
#part 1
#Estimate the intercept coefficient (α) and slope coefficient (β) for each of the ten industry portfolio
#
#parameters setting
filename_industry = "Industry_Portfolios.xlsx"
filename_market = "Market_Portfolio.xlsx"
r_f = 0.13

#industry = pd.read_csv(filename_industry, index_col=0)
industry = pd.read_excel(filename_industry, index_col=0)
industry_cols = industry.columns
#market = pd.read_csv(filename_market, index_col=0)
market = pd.read_excel(filename_market, index_col=0)
market_col = market.columns[0]


#combine two df together
df = pd.merge(market, industry, left_index=True, right_index=True)
results = pd.DataFrame(columns=['Industry', 'Alpha', 'Beta'])


for industry in industry_cols:
    #x
    excess_returns_market = df[market_col] - r_f
    #y
    excess_returns_industry = df[industry] - r_f
    #regress
    #[(Ri-Rf) = alpha + beta*(Rm-Rf)]
    cofficients = np.polyfit(excess_returns_market,excess_returns_industry,deg=1)
    beta = cofficients[0] #slope
    alpha = cofficients[1] #intercept
    
    new_row = pd.DataFrame({'Industry': [industry], 'Alpha': [alpha], 'Beta': [beta]})
    results = pd.concat([results, new_row], ignore_index = True, axis = 0)
results


#part 2.1
#estimated intercept and slope coefficients for the SML
#
mean = pd.DataFrame({"mean":df.mean()})

betas = results.copy()
betas.set_index("Industry", inplace = True)
betas = betas[["Beta"]]

regress = pd.merge(mean, betas, left_index=True, right_index = True, how = "outer")
regress["Beta"][market_col] = 1

#regress returns and beta
#R_p = R_f + (R_m - R_f)*beta_p
#regress
cofficients = np.polyfit(regress["Beta"], regress["mean"],deg=1)
R_f = cofficients[1]#intercept 
risk_premium = cofficients[0]#slope 
print(f"The intercept({R_f}) and slope coefficients({risk_premium}) for the SML")


#part 2.2 plotting
#
#
#parameters setting
x_start = 0
x_end = 2

xlim_start = 0
xlim_end = 2

label1 = "industry portfolios"
label2 = 'SML'

xlabel = "Beta"
ylabel = 'Expected Return(%)'
title = "Security Market Line"

# intercept and slope
#R_p = R_f + (R_m - R_f)*beta_p
intercept = R_f
slope = risk_premium

# create data points
x = np.linspace(x_start, x_end, 1000)
plt.xlim(xlim_start,xlim_end)
y = intercept + slope * x  

plt.scatter(regress["Beta"], regress["mean"], label=label1)
plt.plot(x, y, color='red', label=label2)  

plt.xlabel(xlabel)
plt.ylabel(ylabel)
plt.title(title)

plt.legend()

plt.show()
```

## CAPM intercept and slope coefficients
Briefly explain the economic significance of the intercept and slope coefficients.

**Intercept coefficient from market model regression shows pricing error relative to CAPM.**
>If the intercept coefficient is positive, it suggests that the asset has been earning higher returns than what would be predicted by the CAPM, indicating that the asset might be undervalued. Conversely, a negative intercept implies the asset has been underperforming relative to CAPM expectations, potentially indicating overvaluation.

**Slope coefficient from market model regression shows degree of exposure to market risk.**
>If the slope coefficient (beta) is greater than 1, it signifies that the asset is more volatile than the overall market. A beta less than 1 suggests that the asset is less volatile than the market. Beta provides valuable information to investors about an asset's sensitivity to market movements, helping them assess and manage risk in their portfolios.

# 3.Linear Factor Model

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
#part 1 
#Create a table showing the performance metrics for the ten industry portfolios.
#
#parameters setting
filename_risk_factors = "Risk_Factors.xlsx"
filename_industry_portfolios = "Industry_Portfolios.xlsx"
filename_market_portfolio = "Market_Portfolio.xlsx"

Rm_Rf = "Rm-Rf"
SMB = "SMB"
HML = "HML"

#starting to code
risk_factors = pd.read_excel(filename_risk_factors,index_col=0)
industry_portfolios = pd.read_excel(filename_industry_portfolios, index_col=0)
market_portfolio = pd.read_excel(filename_market_portfolio, index_col=0)

data = pd.merge(market_portfolio, industry_portfolios, left_index=True, right_index=True, how = "inner")
data = pd.merge(data, risk_factors, left_index=True, right_index=True, how="inner")

#calculate risk premium for all industries
data1 = data.copy()
cols = []
for i in industry_portfolios.columns:
    col = f"{i}-Rf"
    cols.append(col)
    data1[col] = data1[i] - data1["Rf"]

#sharpe_ratio E(Ri-Rf)/(var(Ri-Rf))**0.5
metrics = pd.DataFrame(data1[cols].apply(lambda x: np.mean(x)/np.std(x)))
metrics.columns = ["sharpe_ratio"]

#Sortino ratio (using risk-free rate as target) E(Ri-Rt)/(SV(Ri;Rt))**0.5
metrics["Sortino_ratio"] = (
    data1[cols]
    .apply(
        lambda x: 
        np.mean(x)
        /
        np.mean(np.clip(x,None,0)**2)**0.5
    )
)

#Trynor ratio E(Ri-Rf)/Beta_i [(Ri-Rf) = alpha + beta*(Rm-Rf)]
metrics["Trynor_ratio"] = (
    pd.DataFrame(
        data1[cols]
        .apply(lambda y: 
               np.mean(y)
               /
               np.polyfit(data1["Rm-Rf"], y, deg=1)[0]
              )
    )
)

#Jensen's α  alpha = E(Ri-Rf)-Beta_i*E(Rm-Rf) [(Ri-Rf) = alpha + beta*(Rm-Rf)] same alpha
metrics["Jensen_alpha"] = pd.DataFrame(data1[cols].apply(lambda y: np.polyfit(data1["Rm-Rf"], y, deg=1)[1]))

#three-factor alpha
#Ri-Rf = alpha + beta*(Rm-Rf) + gamma*(Rs-Rb) + delta*(Rh-Ri)
data2 = data1.copy()
data2["three_factor_alpha"] = 1
metrics["three_factor_alpha"] = (
    data2[cols].apply(
        lambda y: 
        np.linalg.lstsq(
            data2[["three_factor_alpha", Rm_Rf, SMB, HML]]
            , y)[0][0]
    ) 
)

metrics.index = industry_portfolios.columns


#part 2 plotting
#
#
#sharpe_ratio
plt.bar(metrics.index, metrics["sharpe_ratio"])

for i in range(len(metrics["sharpe_ratio"])):
    plt.text(metrics.index[i], metrics["sharpe_ratio"][i], str(round(metrics["sharpe_ratio"][i],2)), ha='center', va='bottom')
    
plt.title("Sharpe Ratio")
plt.show()



#Sortino Ratio
plt.bar(metrics.index, metrics["Sortino_ratio"])
plt.title("Sortino Ratio")

for i in range(len(metrics["Sortino_ratio"])):
    plt.text(metrics.index[i], metrics["Sortino_ratio"][i], str(round(metrics["Sortino_ratio"][i],2)), ha='center', va='bottom')
    
plt.show()

#Trynor Ratio
plt.bar(metrics.index, metrics["Trynor_ratio"])
plt.title("Trynor Ratio")
plt.ylabel("%")

for i in range(len(metrics["Trynor_ratio"])):
    plt.text(metrics.index[i], metrics["Trynor_ratio"][i], str(round(metrics["Trynor_ratio"][i],2)), ha='center', va='bottom')
    
plt.show()


#Jensen Alpha
plt.bar(metrics.index, metrics["Jensen_alpha"])
plt.title("Jensen Alpha")
plt.ylabel("%")


for i in range(len(metrics["Jensen_alpha"])):
    if metrics["Jensen_alpha"][i] >= 0:
        va = "bottom"
    else:
        va = "top"
    plt.text(metrics.index[i], metrics["Jensen_alpha"][i], str(round(metrics["Jensen_alpha"][i],2)), ha='center', va=va)

plt.show()

#3 Factor Alpha
plt.bar(metrics.index, metrics["three_factor_alpha"])
plt.title("3 Factor Alpha")
plt.ylabel("%")

for i in range(len(metrics["three_factor_alpha"])):
    if metrics["three_factor_alpha"][i] >= 0:
        va = "bottom"
    else:
        va = "top"
    plt.text(metrics.index[i], metrics["three_factor_alpha"][i], str(round(metrics["three_factor_alpha"][i],2)), ha='center', va=va)
    
plt.show()
```

## Three performance ratios

Briefly explain the economic significance of each of the three performance ratios (but not α's).

### sharpe ratio

**Sharpe ratio represents risk premium per unit of total risk:**

**Includes idiosyncratic risk, which penalises individual investments and non-diversified portfolios**

**Implicitly assumes normal returns, so cannot distinguish between asymmetric return distributions with same variance but different skewness**
>**Risk-Adjusted Returns**: The Sharpe Ratio is a vital tool for investors as it quantifies the risk premium earned per unit of total risk taken. This ratio serves as a practical measure to assess the performance of an investment or portfolio, considering both the expected return and the level of risk involved.

>**Incorporating Diversification**: The Sharpe Ratio's consideration of total risk, which includes both systematic (market-related) and idiosyncratic (investment-specific) risk, is of economic significance. It highlights the importance of diversification to manage risk effectively. Individual investments or non-diversified portfolios tend to have lower Sharpe Ratios, signaling the need for risk mitigation through diversification.

>**Normal Return Assumption**: An essential point of consideration is that the Sharpe Ratio implicitly assumes that returns follow a normal distribution. This is where its economic relevance meets a limitation. It means the ratio cannot distinguish between investments with the same variance but differing skewness in their return distributions. In situations where returns exhibit significant asymmetry or fat tails (i.e., non-normal distributions), the Sharpe Ratio may not accurately capture the investment's risk-return profile.

### Sortino ratio

**Sortino ratio represents risk premium per unit of downside risk, which can distinguish between asymmetric return distributions with same variance but different skewness**
>**Focused on Downside Risk**: The Sortino Ratio is a valuable metric for investors because it specifically measures the risk premium earned per unit of downside risk. Unlike the Sharpe Ratio, which considers total risk, the Sortino Ratio concentrates on the potential for losses and unfavorable outcomes. This makes it particularly relevant for risk-averse investors who are more concerned with avoiding losses.

>**Distinguishing Asymmetric Returns**: One of the primary economic advantages of the Sortino Ratio is its ability to distinguish between investment opportunities with asymmetric return distributions. It can differentiate between assets that have the same variance but exhibit different skewness in their return distributions. This is crucial because it recognizes that investors are often more concerned about downside risk (losses) than overall risk.

>**Risk Management and Portfolio Selection**: The Sortino Ratio is a practical tool for assessing and managing risk in investment portfolios. It helps investors make informed decisions about asset allocation and risk control by emphasizing the importance of minimizing downside risk while seeking to maximize returns.

>**Aligning with Investor Preferences**: For investors who are more risk-averse and prioritize capital protection, the Sortino Ratio offers a risk-adjusted performance measure that better aligns with their preferences. It provides a more focused perspective on downside risk, helping them select investments that meet their specific risk tolerance and financial goals.

### Treynor ratio

**Treynor ratio represents risk premium per unit of market risk, which ignores idiosyncratic risk as well as other types of systematic risk**

>**Market Risk Assessment**: The Treynor Ratio is a critical tool for investors because it measures the risk premium earned per unit of market risk. This ratio specifically focuses on systematic or market risk, which is the portion of an asset's risk that cannot be eliminated through diversification.

>**Idiosyncratic Risk Exclusion**: One of the key economic benefits of the Treynor Ratio is its exclusion of idiosyncratic risk, which is the risk that is specific to individual investments. By disregarding this idiosyncratic risk, the Treynor Ratio provides a more concentrated evaluation of an asset's performance concerning market-related factors.

>**Comparative Analysis**: The Treynor Ratio enables investors to compare different investments or portfolios based on their exposure to systematic risk. This makes it particularly relevant for investors who are primarily concerned with the broader market movements rather than specific attributes of individual assets.

>**Risk-Return Efficiency**: Investors use the Treynor Ratio to assess the efficiency of an investment or portfolio in terms of risk-adjusted returns. It helps in evaluating whether the returns generated are commensurate with the level of systematic risk taken, allowing for more informed investment decisions.

>**Investment Strategy**: The Treynor Ratio is instrumental in shaping investment strategies that aim to optimize the risk-return tradeoff. By focusing on market risk, it helps investors tailor their portfolios to match their risk preferences and financial goals.

### Jensen's Alpha

**For passive portfolio, represents pricing error relative to CAPM**

**For active portfolio, represents abnormal mean return after adjusting for exposure to market risk, due to fund manager's ability to identify underpriced or overpriced assets**

**Can extend to multiple sources of systematic risk by using intercept coefficient from multi-factor pricing model**

>**Performance Evaluation**: Investors and fund managers use Jensen's Alpha as a tool for evaluating the skill and performance of portfolio managers. A positive alpha suggests that the manager has added value beyond what could be attributed to market risk exposure, making them a potentially attractive choice for investors seeking active management.

>**Investment Decision-Making**: The economic significance of Jensen's Alpha is also in helping investors make informed investment decisions. A positive alpha may indicate that an investment or fund is delivering superior risk-adjusted returns, potentially justifying its inclusion in a portfolio.

### 3 factor alpha

**If model is correct, then intercept coefficient should be zero for any individual asset or passive portfolio**

**Hence interpret intercept coefficient as pricing error, relative to F–F three-factor model**

>**Pricing Error Assessment**: The Three-Factor Alpha is economically significant because it plays a crucial role in assessing pricing errors in the context of the Fama-French Three-Factor Model. In an accurate model, the intercept coefficient, which represents the alpha, should be zero for any individual asset or passive portfolio.

>**Alpha as an Indicator**: The alpha, when not zero, is interpreted as a measure of pricing error relative to the Three-Factor Model. A positive alpha suggests that the asset or portfolio has outperformed what would be expected based on the model, potentially indicating undervaluation. Conversely, a negative alpha suggests underperformance, possibly indicating overvaluation.

>**Portfolio Evaluation**: For investors, the Three-Factor Alpha provides a tool for evaluating the performance of their portfolios in a more sophisticated way. It allows them to assess whether the returns generated are truly attributable to the market, size, and value risk factors, or if there are other unexplained sources of return that might need further investigation.

>**Investment Strategy**: The Three-Factor Alpha informs investment strategies by helping investors identify assets or portfolios that may offer alpha and potentially superior risk-adjusted returns. Investors can make decisions based on whether they believe an asset's alpha is sustainable or merely a result of pricing error.

## H-J bound

Explain (in words, without using any mathematical  equations or formulas) the economic significance of this result.

**H–J bound: volatility ratio of pricing kernel must exceed Sharpe ratio of 0.4 for US stock market**

**For investor with power utility of consumption, investor's coefficient of relative risk aversion (γ) affects volatility of pricing kernel ⇒ H–J bound sets lower bound on γ**

**H–J bound is satisfied for reasonable value of γ when probability distribution of consumption growth includes rare disasters ⇒ no equity premium puzzle**

>**Volatility Ratio and Sharpe Ratio**: The H-J bound states that the volatility ratio of the pricing kernel must exceed a certain threshold, specifically the Sharpe ratio of 0.4 for the US stock market. The volatility ratio measures the risk inherent in asset prices, while the Sharpe ratio measures the excess return per unit of risk. When the volatility ratio falls short of this threshold, it indicates that asset prices may not adequately compensate investors for the risks they are exposed to in the market.

>**Investor Behavior and Risk Aversion**: The investor's coefficient of relative risk aversion (γ) plays a crucial role in determining how they make investment decisions. Higher values of γ indicate greater aversion to risk, while lower values suggest a higher tolerance for risk. The H-J bound tells us that γ affects the volatility of the pricing kernel. In other words, the investor's risk aversion influences the way assets are priced and their associated risk levels

>**Setting a Lower Bound on γ**: The H-J bound effectively sets a lower limit on the acceptable level of γ. It implies that γ should be at least at a certain level to ensure that asset prices are consistent with market realities. This is important because it helps in understanding the minimum degree of risk aversion that investors should exhibit to justify the prevailing asset prices in the market.

>**Rare Disasters and Equity Premium Puzzle**: One of the notable implications of the H-J bound is that it can help resolve the "equity premium puzzle." The equity premium puzzle refers to the question of why investors appear to demand higher returns from investing in stocks compared to safer assets, given the observed risk and return characteristics of the stock market. The H-J bound suggests that this puzzle can be resolved when the probability distribution of consumption growth includes rare disasters. In other words, when investors account for the possibility of catastrophic events impacting their consumption, they may require higher returns from riskier assets like stocks to compensate for that added risk.

```python
#parameters setting
lng_intercept = 0.02
lng_slope = 0.02
simulate_time = int(1e4)

Rf = 1.0303

penalize_coefficient = 2
##############################
#g = e^(0.02 + 0.02*epsilon)
#lng = 0.02 + 0.02*epsilon
gs = np.exp(lng_intercept+lng_slope*np.random.normal(0,1,simulate_time))

def cal_E(x,gs):
    summation = 0
    for g in gs:
        if x*g >= Rf:
            summation += x*g - Rf
        else:
            summation += penalize_coefficient*(x*g - Rf)
    mean = summation/len(gs)
    return mean

#parameters setting
#e(x) = 0.99*b_0*E(v(x*g)) + 0.99*x - 1
def cal_x(b0, x1, x2):
    
    x = (x1+x2)/2
    
    while(abs(0.99*b0*cal_E(x, gs) + 0.99*x - 1))>(1e-5):
        if (0.99*b0*cal_E(x, gs) + 0.99*x - 1)>0:
            x2 = x
        else:
            x1 = x
        x = (x1+x2)/2
        
    return x      
###########################################

#parameters setting
b0_start = 0
b0_end = 10.1
b0_increment = 0.1

x1 = 1
x2 = 1.1

Rm_coef = 0.0202
####################################################

df_ans = pd.DataFrame([np.arange(b0_start,b0_end,b0_increment)], index=["b0"]).T

lst_x = []
for b0 in df_ans["b0"]:
    lst_x.append(cal_x(b0, x1, x2))
    
df_ans["x"] = lst_x

df_ans["price_dividend_ratio"] = 1/(df_ans["x"]-1)

#E(Rm) = E(xg) = x*e^(0.0202)
df_ans["E_Rm"] = df_ans["x"]*np.exp(Rm_coef)

df_ans["ERP"] = df_ans["E_Rm"] - Rf

#parameters setting
xlabel = "b0"
ylabel = "price-divident ratio"
title = "price-dividend ratio vs b0"
###################################################
#plt.figure(figsize=(19,19))
plt.plot(df_ans["b0"], df_ans["price_dividend_ratio"])
plt.xlabel(xlabel)
plt.ylabel(ylabel)
#plt.xticks(df_ans["b0"],rotation=45,fontsize=9)
plt.title(title)
plt.show()

#parameters setting
xlabel = "b0"
ylabel = "equity premium"
title = "equity premium vs b0"
############################################
plt.xlabel(xlabel)
plt.ylabel(ylabel)
plt.title(title)
plt.plot(df_ans["b0"],df_ans["ERP"])
plt.show()
```

Briefly explain main characteristics of v(∙) (which is utility function to measure utility from recent financial gain or loss), as well as economic significance of b0 and λ.

**Utility function to measure utility from recent financial gain or loss is based on prospect theory, where financial gain or loss is measured relative to reference level based on risk-free rate**

>**Reference Point**: One of the primary characteristics of v(∙) is that it evaluates financial gains or losses relative to a reference level. This reference level is often based on the risk-free rate or some other relevant benchmark. The reference point serves as a point of comparison for assessing the value of recent financial changes.

>**Loss Aversion**: Prospect theory, which underlies v(∙), highlights the idea that individuals tend to be more sensitive to financial losses than gains. This is a key characteristic of the utility function. It means that the psychological impact of losing a certain amount is greater than the psychological impact of gaining the same amount. This asymmetry in how individuals perceive gains and losses is a fundamental aspect of v(∙).

**Investor is more sensitive to financial loss, and λ determines degree of loss aversion**

>**Loss Aversion (λ)**: The parameter λ, also known as the loss aversion coefficient, is a critical element of v(∙). It quantifies the degree of loss aversion exhibited by an individual. A higher value of λ implies stronger loss aversion, meaning that losses have a more significant impact on an individual's overall utility compared to gains of the same magnitude. Understanding λ is essential for modeling how people make financial decisions. It can explain why individuals may be more risk-averse when faced with potential losses and why they tend to avoid investments that could lead to significant losses.

**b0 determines extent to which utility from recent financial gain or loss contributes to investor's overall lifetime utility**
>**Contribution to Lifetime Utility (b0)**: The parameter b0 is related to how much weight is given to recent financial gains or losses in an individual's overall lifetime utility function. It determines the extent to which these recent financial experiences influence decision-making. A higher value of b0 indicates that recent gains or losses have a greater impact on an individual's utility, affecting their choices in investment, consumption, and risk-taking. The economic significance of b0 lies in understanding how individuals weigh recent financial events when making choices. It can help in predicting investor behavior and preferences.

### Black-Litterman Model

### problem
>**Challenges in Estimating Mean Returns**: The Black-Litterman model addresses a significant econometric issue in portfolio construction—the difficulty of accurately estimating mean returns for assets. The sample mean return is commonly used, but it has limitations, especially when dealing with a small number of observations or volatile returns.

>**Standard Error of Sample Mean**: The sample mean has a standard error that is inversely proportional to the square root of the number of observations (m). With returns being volatile and data often limited, the estimate of the mean return tends to have a large standard error.

>**Efficient Frontier Sensitivity**: Small changes in the vector of mean returns can result in substantial changes in the location of the efficient frontier. This sensitivity poses challenges for investors relying solely on historical mean returns to construct optimal portfolios.


### solution
>**Implied Sample Risk Premiums**: The Black-Litterman model addresses these challenges by using the observed weights of risky assets in the market portfolio to determine implied sample risk premiums. This approach helps in deriving more robust estimates of expected returns, considering the market's collective wisdom.

>**Calibration of λ**: The model introduces a parameter (λ) that needs to be calibrated. This is often done using market risk premiums or Sharpe ratios, providing a mechanism to balance the impact of views against the implied market expectations.

### strength
>**Incorporating Investor's Views**: One unique feature of the Black-Litterman model is its ability to incorporate investor views on expected returns. These views can be expressed as absolute expectations or relative expectations between assets. This inclusion allows investors to integrate their subjective beliefs into the optimization process.

>**Bayesian Framework**: The model uses Bayes' theorem to systematically incorporate investor views into the distribution of excess returns. This Bayesian approach combines the market equilibrium implied returns with the investor's views, resulting in a blended expected return distribution.

>**Enhanced Portfolio Construction**: By addressing the challenges in estimating mean returns and incorporating investor views, the Black-Litterman model offers a more robust and practical framework for portfolio construction. It provides a structured way to leverage both market equilibrium and investor insights, leading to portfolios that better reflect the investor's risk-return preferences.
